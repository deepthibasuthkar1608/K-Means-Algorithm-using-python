```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the CSV data file
path= "/content/Wholesale_customers_data[1].csv"
data = pd.read_csv(path)

# Extract numerical data
X = data.iloc[:, 1:].values  # Assuming the numerical data starts from the second column

# Preprocessing: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Initialize KMeans
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)


# Fit KMeans
kmeans.fit(X_scaled)

# Get cluster labels
labels = kmeans.labels_

# Get cluster centroids
centroids = kmeans.cluster_centers_

# Add cluster labels to the original dataframe
data['Cluster'] = labels

# Analyze the results
cluster_counts = data['Cluster'].value_counts()
print("Cluster Counts:")
print(cluster_counts)
print("\nCluster Centroids:")
print(centroids)
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the CSV data into a DataFrame
path= "/content/Wholesale_customers_data[1].csv"
data = pd.read_csv(path)

# Select the features for clustering
X = data[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']].values

# Data preprocessing (standardization)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Choose the number of clusters (k)
k = 5

# Perform K-means clustering
kmeans = KMeans(n_clusters=k, random_state=42)
y_predict = kmeans.fit_predict(X_scaled)

# Add the cluster labels to the DataFrame
data['cluster'] = y_predict

# Get the cluster centroids
centroids = kmeans.cluster_centers_

# Print the centroids
print("Cluster centroids:")
for i, centroid in enumerate(centroids):
    print(f"Cluster {i+1}: {centroid}")

# Visualize the clusters
plt.figure(figsize=(12, 8))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_predict, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', label='Centroids')
plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.title('K-means Clustering of Wholesale Customers')
plt.legend()
plt.show()
```

```python
import numpy as nm
import matplotlib.pyplot as mtp
import pandas as pd

# Load the CSV data into a DataFrame
path = "/content/Wholesale_customers_data[1].csv"
data = pd.read_csv(path)

x = data.iloc[:, [3, 4]].values  # Use 'data' instead of 'dataset'

# Finding optimal number of clusters using the elbow method
from sklearn.cluster import KMeans
wcss_list = []  # Initializing the list for the values of WCSS

# Using for loop for iterations from 1 to 10.
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)
mtp.plot(range(1, 11), wcss_list)
mtp.title('The Elbow Method Graph')
mtp.xlabel('Number of clusters (k)')
mtp.ylabel('WCSS (Within-Cluster Sum of Squares)')
mtp.show()

# Training the K-means model on the dataset
kmeans = KMeans(n_clusters=8, init='k-means++', random_state= 42)
y_predict = kmeans.fit_predict(x)

# Visualizing the clusters
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'black', 'purple', 'orange']
for i in range(8):
    mtp.scatter(x[y_predict == i, 0], x[y_predict == i, 1], s = 100, c = colors[i], label = f'Cluster {i+1}')

mtp.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroid')
mtp.title('Clusters of customers')
mtp.xlabel('X axis')
mtp.ylabel('Y axis')
mtp.legend()
mtp.show()
```
